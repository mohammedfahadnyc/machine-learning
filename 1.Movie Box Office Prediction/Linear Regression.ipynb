## 1. Creating X,Y (Dataframe) Using Pandas

```python
import pandas
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data
```

```python
data.describe()    #finding mean,median etc
```

```python

from pandas import DataFrame
X = DataFrame(data,columns=['production_budget_usd'])
```

```python
Y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
X,Y
```

## 2. Plotting using matplotlib

```python

#Plotting using matplotlib
from matplotlib import pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title("Budget vs Revenue")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

## 3. Creating Regression Model

```python
#Creating the regression model
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X,Y)
regression.coef_, regression.intercept_
```

```python
#Creating the line of best fit
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title("Budget vs Revenue")
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.plot(X,regression.predict(X),color='red',linewidth=3)
plt.show()


```

```python
#Eavluating the model, note: We are evaluating using the data we used to train the model, we should also test on new data
regression.score(X,Y)
```

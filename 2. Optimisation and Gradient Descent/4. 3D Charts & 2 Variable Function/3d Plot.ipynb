## Data visualization - 3D Charts and 2 Variable Function

$$ f(x,y) = \frac {1} {{3^{-x^2 - y^2}}+1} $$

```python
def f(x,y):
    r = 3**(-x**2 - y**2)
    return 1/(r+1)
```

## Populate x,y coordinates

```python
import numpy as np
x_4 = np.linspace(-2,2,200)
y_4 = np.linspace(-2,2,200)
x_4.shape
```

## Create the 3D Plot

We need a figure and axes details

```python
# Creating figure object and adding 3d axes to it
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(projection='3d')

#drawing the surface, plot_surface takes x,y,z which must be 2d, 
#but our x_4,y_4 are 1D vector, so convert to 2D First
#For color, import cm from matplotlib, and plot_surface take cmap, cmap=cm.coolwarm etc
from matplotlib import cm
x_4,y_4 = np.meshgrid(x_4,y_4)
ax.plot_surface(x_4,y_4,f(x_4,y_4),cmap=cm.coolwarm,alpha=0.6)
```
